```python
import pandas as pd
```

```python
wnv = pd.read_csv('wnv_aggr_final.csv') #The WNV data given by Dr.Aiman
se = pd.read_csv('se_data.csv') #The Socio-Economic data that was cleaned by Vinu
rc = pd.read_csv('race_data.csv')
rohan = pd.read_csv("./Poverty/ACS_12_5YR_S1701/ACS_12_5YR_S1701_with_ann.csv",encoding='ISO-8859-1',low_memory=False)
```

```python
#Getting the idea of the wnv dataset
print(wnv.columns)
print("\nShape: ", wnv.shape)
print("\nYears: ", wnv.YEAR.unique())
print("\nCounties: ", len(wnv.GEOID10.unique()))
wnv.head(5)
```

```python
print(se.columns)
print("\nShape: ", se.shape)
print("\nYears: ", se.YEAR.unique())
print("\nCounties: ", len(se.STCOU.unique()))
se.head(5)
```

```python
se['Calculated_population'] = (se['Poverty_number_of_people'] * 100)/se['Poverty_percent_of_people']
pd.options.display.float_format = '{:.2f}'.format
se.head(5)
```

```python
print(rc.columns)
print("\nShape: ", rc.shape)
print("\nYears: ", rc.YEAR.unique())
print("\nCounties: ", len(rc.STCOU.unique()))
rc.head(5)
```

```python
rc['Total_number'] = rc['Resident_population_White_alone'] + rc['Resident_population_Black_alone'] + rc['Resident_population_American_Indian_and_Alaska_Native_alone'] + rc['Resident_population_Asian_alone'] + rc['Resident_population_Native_Hawaiian_and_Other_Pacific_Islander_alone'] + rc['Resident_population_Two_or_more_races'] 
rc['Total_percent'] = rc['Resident_population_White_alone_percent'] + rc['Resident_population_Black_alone_percent'] + rc['Resident_population_American_Indian_and_Alaska_Native_alone_percent'] + rc['Resident_population_Asian_alone_percent'] + rc['Resident_population_Native_Hawaiian_and_Other_Pacific_Islander_alone_percent'] + rc['Resident_population_Two_or_more_races_percent'] 
rc.head(5)
```

```python
for i in rohan.iloc[0]:
    print(i)
```

```python
wnv_2000 = wnv[wnv['YEAR'] == 2000]
se_2000 = se[se['YEAR'] == 2000]
rc_2000 = rc[rc['YEAR'] == 2000]
```

```python
print(wnv_2000.shape)
wnv_2000[['YEAR' , 'GEOID10', 'CTYNAME', 'POPESTIMATE']].head(5)
```

```python
se_2000[['YEAR','STCOU','Areaname','Calculated_population']].head(6)
```

```python
print(rc_2000.shape)
rc_2000[['YEAR','STCOU','Areaname','Total_number']].head(6)
```

```python
result = pd.merge(wnv_2000, rc_2000, how='inner', left_on='GEOID10', right_on='STCOU')
```

```python
print(result.shape)
result.columns
```

```python
result = result[['YEAR_x', 'GEOID10', 'CTYNAME', 'POPESTIMATE','Total_number']]
result["POPESTIMATE"] = pd.to_numeric(result["POPESTIMATE"])
result.head(5)
```

```python
result.dtypes
```

```python
result['Error'] = (result["POPESTIMATE"] - result["Total_number"]).abs()
result['Percentage_error'] = (result["Error"]/result["POPESTIMATE"])*100

print("Values with greater than 10% error: \n\n",result[result.Percentage_error > 10.0 ].count())

#Sorting by percentage error
result.sort_values(by=['Percentage_error'], ascending=False)
```

```python
print("Values with greater than 5% error: \n\n",result[result.Percentage_error > 2.0 ].count())

#Sorting by percentage error
result.sort_values(by=['Percentage_error'], ascending=False)
```
