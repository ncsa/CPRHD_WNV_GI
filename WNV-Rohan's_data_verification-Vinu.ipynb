```python
data = data.iloc[:, : 9]
data.columns
```

```python
data = data[['GEO.id2', 'GEO.display-label', 'HC01_EST_VC01', 'HC02_EST_VC01','HC03_EST_VC01']]
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data.head(5)
```

```python
data.columns
```

```python
data["Total; Estimate; Population for whom poverty status is determined"] = pd.to_numeric(data["Total; Estimate; Population for whom poverty status is determined"])
data["Below poverty level; Estimate; Population for whom poverty status is determined"] = pd.to_numeric(data["Below poverty level; Estimate; Population for whom poverty status is determined"])
data["Percent below poverty level; Estimate; Population for whom poverty status is determined"] = pd.to_numeric(data["Percent below poverty level; Estimate; Population for whom poverty status is determined"])
data.shape
```

```python
data["Calculated_Population"] = (data["Below poverty level; Estimate; Population for whom poverty status is determined"]
                                 /data["Percent below poverty level; Estimate; Population for whom poverty status is determined"])*100
```

```python
data.head(5)
```

```python
data['is_calculated_pop_greater'] = (data['Total; Estimate; Population for whom poverty status is determined'] < data['Calculated_Population']).astype(bool)
```

```python
data.head(5)
```

```python
print("Count where calculated population is greater:")
data.is_calculated_pop_greater.astype(bool).sum(axis=0)
```

```python
print("Count where calculated population is lesser:")
zero = data[data['is_calculated_pop_greater'] == False]
zero.shape[0]
```
